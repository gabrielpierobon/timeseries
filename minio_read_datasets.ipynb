```python
import boto3
from botocore.client import Config
import pandas as pd
import io
```

```python
# MinIO credentials and endpoint
ACCESS_KEY = 'minio'
SECRET_KEY = 'minio123'
MINIO_URL = 'http://172.23.0.6:9000'
BUCKET_NAME = 'timeseriesbucket'  # Replace with the name of the bucket you created
```

```python
# Initialize a session using Amazon S3 service and endpoint
session = boto3.session.Session()
s3_client = session.client('s3',
                           endpoint_url=MINIO_URL,
                           aws_access_key_id=ACCESS_KEY,
                           aws_secret_access_key=SECRET_KEY,
                           config=Config(signature_version='s3v4'))

# For example, to list objects in your bucket:
for obj in s3_client.list_objects(Bucket=BUCKET_NAME)['Contents']:
    print(obj['Key'])
```

```python
# Fetch the CSV file from MinIO
csv_obj = s3_client.get_object(Bucket=BUCKET_NAME, Key="inflacion.csv")
```

```python
csv_body = csv_obj['Body'].read()
```
